데이터 불러오기 & 살펴보기

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
car_df = pd.read_csv('data/cars.csv')
brand_df = pd.read_csv('data/brand.csv')
```

```python
car_df.head()
```

```python
brand_df.head()
```

car_df 와  brand_df  합치기

```python
# Q. car_df와 brand_df를 합치기 위해, car_df에 brand 컬럼을 추가.  column 추가 car_df['brand']-- 요부분
# title 값의 첫 번째 단어만 추출하여 brand 컬럼에 저장.    str.split 쓰고,  exand=True로 data frame으로 만들어주기

car_df['brand']=car_df['title'].str.split(expand=True)[0]
```

```python
#brand가 추가되었는 지 확인
car_df.head()
```

```python
# Q. brand_df의 title 컬럼에 저장된 값들을 대문자로 변환 

brand_df['title']=brand_df['title'].str.upper()
```

```python
#brand data set 확인
brand_df.head()
```

```python
# Q. car_df의 brand 컬럼과 brand_df의 title 컬럼을 기준으로 데이터를 merge

car_df=car_df.merge(brand_df, left_on='brand', right_on='title', how='left')
```

```python
car_df.head()
```

```python
# Q. car_df에서 중복되는 컬럼 title_y 제거

car_df=car_df.drop(columns=['title_y'])
```

```python
# Q. car_df에서 title_x 컬럼의 이름을 title로 변경

car_df=car_df.rename(columns={'title_x': 'title'})
```

```python
# column 확인
car_df.head()
```

```python
# car_df의 복사본 bonus_df를 생성합니다.

bonus_df = car_df.copy()
```

Category 형 변수를 숫자로 변환하기

```python
# car_df의 컬럼 각각에 대해 자료형(data type)을 살펴봅니다.

car_df.info()
```

```python
#data table 확인
car_df.head()
```

```python
# Q. Engine 컬럼의 데이터에서 L 문자를 제거하여 숫자 부분만 저장합니다.

car_df['Engine']=car_df['Engine'].str.replace('L', '')
```

```python
# Q. Emission Class 컬럼에서도 숫자 부분만 추출하여 저장합니다.

car_df['Emission Class']=car_df['Emission Class'].str.replace('Euro', '')
```

```python
# Q. to_numeric 메서드를 사용하여 Engine 컬럼의 자료형을 숫자 형태로 변환합니다.

car_df['Engine']=pd.to_numeric(car_df['Engine'])
```

```python
# Q. to_numeric 메서드를 사용하여  Emission Class 컬럼의 자료형을 숫자 형태로 변환합니다.

car_df['Emission Class']=pd.to_numeric(car_df['Emission Class'])
```

```python
#Engine 과  Emission Class의 data type이 실수로 바뀌었는 지 확인
car_df.info()
```

```python
#describe를 통해  통계값으로도 확인.
car_df.describe()
```

```python
# car_df의 컬럼별로 결측치의 비율을 확인
#Service History 85% > Previous Owner 38% > Emission Class 2.3% > Engine 1% > Seat 0.9% > Door 0.6%

car_df.isna().mean()
```

```python
# Q. Service history 컬럼의 결측치를 'Unknown'으로 채우기.   missing value가 너무 많으니, 삭제할 수 없음.

car_df['Service history']=car_df['Service history'].fillna('Unknown')
```

```python
car_df.groupby('Service history')['Price'].mean()
```

```python
# Engine 컬럼의 missing value 확인

car_df[car_df['Engine'].isna()]
```

```python
# 결측치가 포함된 컬럼의 개수를 데이터마다 계산하여 na_values 컬럼에 저장.   column을 새로 만들어서 확인하고 나중에 지움.
# column 값이 많이 missing 되어 있으면, 그 row는 제거 (의미가 없으므로)

car_df['na_values'] = car_df.isna().sum(axis = 1)
```
